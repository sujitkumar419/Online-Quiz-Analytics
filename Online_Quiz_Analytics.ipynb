```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

rows = 200

data = {
    "Student_ID": [f"S{100+i}" for i in range(rows)],
    "Student_Name": [random.choice(["Sujit ", "Priya", "Amit", "Neha", "Rohit", "Anjali", "Karan", "Sneha"]) for _ in range(rows)],
    "Age": np.random.choice([16,17,18,19,20,21,120,5], rows),  
    "Gender": np.random.choice(["Male", "male ", "M", "FEMALE", "f"], rows),
    "City": np.random.choice(["Delhi", "mumbai", "KOLKATA", "Bangalore", "chennai"], rows),
    "Quiz_ID": np.random.choice(["QZ01", "QZ02", "QZ03"], rows),
    "Score": np.random.choice([np.nan, "N/A", 45, 55, 65, 75, 85, 95], rows),
    "Total_Questions": 100,
    "Correct_Answers": np.random.randint(40, 100, rows),
    "Incorrect_Answers": np.random.randint(0, 60, rows),
    "Time_Taken_Minutes": np.random.choice([40,50,60,"forty",70], rows),
    "Attempt_Number": np.random.choice([1,2,3], rows),
    "Device_Type": np.random.choice(["Mobile", "Laptop", "Tablet"], rows),
    "Internet_Speed": np.random.choice(["Fast", "slow", "Medium", "FAST "], rows),
    "Submission_Status": np.random.choice(["Submitted", "submitted", "Not Submitted"], rows),
    "Feedback_Rating": np.random.choice([1,2,3,4,5,np.nan], rows),
    "Passed": np.random.choice(["Yes", "No", "Y", "N", "pass", "fail"], rows),
    "Country": "India"
}

df = pd.DataFrame(data)


df = pd.concat([df, df.iloc[5:10]])

df.to_csv("online_quiz_dirty.csv", index=False)

print("Dirty dataset created successfully!")
```

# Clean Gender Column

```python
df["Gender"] =df["Gender"].str.strip().str.lower()
df["Gender"] = df["Gender"].replace({
    "male": "Male",
    "m": "Male",
    "female": "Female",
    "f": "Female"
})
```

```python
df.head()
```

# Clean Score Column

```python
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Score"].fillna(df["Score"].mean(), inplace=True)
                            
```

```python
df.head()
```

# Clean Time Column

```python
df["Time_Taken_Minutes"] = df["Time_Taken_Minutes"].replace("forty", 40)
df["Time_Taken_Minutes"] = pd.to_numeric(df["Time_Taken_Minutes"], errors="coerce")
```

```python
df.head()
```

# Remove Unrealistic Ages

```python
df = df[(df["Age"] >= 15) & (df["Age"] <=25)]
```

```python
df.head()
```

# Clean City Column

```python
df["City"] = df["City"].str.strip().str.title()
```

```python
df.head()
```

# Standarize Passed Column

```python
df["Passed"] = df["Passed"].str.lower()

df["Passed"] = df["Passed"].replace({
    "yes": "Pass",
    "y": "Pass",
    "pass": "Pass",
    "no": "Fail",
    "n": "Fail",
    "fail": "Fail"
})
```

```python
df.head()
```

```python
df["Feedback_Rating"].fillna(df["Feedback_Rating"].median(), inplace=True)
```

# Create Performance Category

```python
def performance(score):
    if score >= 80:
        return "High Performer"
    elif score >= 60:
        return "Average Performer"
    else:
        return "At Risk"

df["Performance_Category"] = df["Score"].apply(performance)
```

```python
sns.countplot(x="Performance_Category", data=df)
plt.title("Performance Categories")
plt.show()
```
